```python
import httpx
import csv
```

```python
with open('suscess1.txt','rt') as f: 
   cr = csv.reader(f)
   for row in cr:
        try:
            r = httpx.post('http://127.0.0.1:8000/pic/beauties/', data={'address_web':row[-1]})
            print(r.json()["id"])
        except:
            with open('import_fail.txt','a') as fail:
                fail.write(row[0]+ ", " + row[1] + "\n")
```
